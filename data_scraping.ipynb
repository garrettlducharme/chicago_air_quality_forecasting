This notebook will be used to extract the hourly meteorological collected from the S.H Bell Chicago air monitoring site. We'll read in the csv and xlsx files as data frames, and then save them locally as csvs with more easily identifiable names. Data will be saved in the 'data/meteorological' folder. Later, this data will be merged with the PM10 concentration readings that were also taken from the S.H Bell monitoring site. 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.epa.gov/il/sh-bell-chicago-air-monitoring-data")
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
url_head = 'https://www.epa.gov'
items = soup.find_all('table')[3].find('tbody').select('a')

data_urls = [url_head + item['href'] for item in items]
```

```python
for data_url in data_urls[1:]:
    file_name = data_url.split('/')[-1].split('.')[0]
    temp_df = pd.read_excel(data_url)
    temp_df.to_csv(f'data/meteorological/{file_name}.csv')
```
